```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
#from sklearn.metrics import accuracy_score,precision_score,\
#recall_score,confusion_matrix,roc_auc_score,average_precision_score

from sklearn.metrics import confusion_matrix,accuracy_score,roc_auc_score,precision_score,\
recall_score,f1_score,roc_curve,cohen_kappa_score,matthews_corrcoef,classification_report

#from utils_multiclass_classifier import *
import seaborn as sns
```

```python
project = "NCI"
predictor = "retrain_mean"

n_class = 10
norm_type = "true"
```

```python
#df = pd.read_csv(f"{project}_raw_demo_HnE_scores.csv")
df = pd.read_csv(f"results_{project}/{predictor}_score.csv")
df
```

```python
df_dict = pd.read_csv(f"../10methylation_meta/NCI-demo_dict_{n_class}class.csv")
class_names = df_dict["DBTA_name"].values

actual_names = df["DBTA_name"].values
pred_names = df["top1_pred_name"].values

acc = np.mean(actual_names == pred_names)
print(acc)
```

```python
#cfs_matrix = confusion_matrix(actual_names, pred_names,labels=class_names).T
#cfs_matrix = confusion_matrix(actual_names, pred_names,labels=class_names,normalize="pred").T
#cfs_matrix = confusion_matrix(actual_names, pred_names,labels=class_names,normalize="true").T

cfs_matrix = confusion_matrix(actual_names, pred_names,labels=class_names,normalize=norm_type).T
#cfs_matrix = (cfs_matrix*100).astype(int)

df_cfs = pd.DataFrame(cfs_matrix, index=class_names, columns=class_names)
df_cfs
```

```python
nx,ny = 1,1
fig, ax = plt.subplots(ny,nx,figsize=(nx*6,ny*6))

##----------------------------------
## confusion matrix
fontsize=8
#heatmap = sns.heatmap(df_cfs, ax=ax, cmap='Blues', fmt='d', annot=True, annot_kws={"size": fontsize+4},
#          linewidths=2, linecolor='black', cbar=False)
heatmap = sns.heatmap(df_cfs, ax=ax, cmap='Blues', fmt='.2f', annot=True, annot_kws={"size": fontsize+4},
          linewidths=0.5, linecolor='green', cbar=False)

ax.set_xlabel('Actual Subtypes', fontsize=14, color='black')
ax.set_ylabel('Predicted Subtypes', fontsize=14, color='black')
ax.set_title(f"{project}, acc: {round(acc,3)}",fontsize=15)

plt.tight_layout(h_pad=1, w_pad= 1.5)
plt.savefig(f"cfs_{project}_{predictor}.pdf", format='pdf', dpi=50)
#plt.close()
```
