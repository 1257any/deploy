```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
#from sklearn.metrics import accuracy_score,precision_score,\
#recall_score,confusion_matrix,roc_auc_score,average_precision_score

from sklearn.metrics import confusion_matrix,accuracy_score,roc_auc_score,precision_score,\
recall_score,f1_score,roc_curve,cohen_kappa_score,matthews_corrcoef,classification_report,auc

from utils_multiclass_classifier import *
import seaborn as sns

from sklearn.preprocessing import LabelBinarizer
```

```python
project = "NCI"

predictor = "demo"
n_class = 10

result_dir = f"results_{project}/"
```

```python
df = pd.read_csv(f"{result_dir}{predictor}_scores.csv")
df
```

```python
df_dict = pd.read_csv(f"../10methylation_meta/NCI_dict_{n_class}class.csv")
df_dict
```

```python
class_names = df_dict["DBTA_class"].values
n_class = len(class_names)
print("n_class:", n_class)

pred_scores = df[class_names].values
actual_names = df["DBTA_class"].values
print(actual_names)
```

```python
actual_idxs = np.array([np.argwhere(class_names == x)[0][0] for x in actual_names])

pred_idxs = np.argmax(pred_scores, axis=1)
pred_names = class_names[pred_idxs]
```

```python
## AUC of
actual_onehot = LabelBinarizer().fit_transform(actual_idxs)

fpr, tpr, _ = roc_curve(actual_onehot.ravel(), pred_scores.ravel())

auc_mean = auc(fpr, tpr)
print("auc_mean:", auc_mean)
df_mean = pd.DataFrame({"auc": [auc_mean]})

auc_each = roc_auc_score(actual_idxs, pred_scores, average=None, multi_class="ovr")
df_dict["auc"] = auc_each

df_roc = pd.DataFrame({"fpr": fpr, "tpr": tpr})
```

```python
## precision and recall of each label
precision_each = precision_score(actual_names, pred_names, labels=class_names, average=None)
recall_each = recall_score(actual_names, pred_names, labels=class_names, average=None)
f1score = f1_score(actual_names, pred_names, labels=class_names, average=None)

df_dict["precision"] = precision_each
df_dict["recall"] = recall_each
df_dict["f1score"] = f1score
```

```python
## topk accuracy
topk_list = [1,2,3]
for topk in topk_list:
    
    topk_pred_names,all_topk_pred_names = find_topk_pred_names(actual_names,pred_scores,class_names,topk)

    ## acc
    acc_mean = np.mean(actual_names == topk_pred_names)
    print(acc_mean)

    cfs_matrix = confusion_matrix(actual_names, topk_pred_names,labels=class_names).T
    df_cfs = pd.DataFrame(cfs_matrix, index=class_names, columns=class_names)

    TP = np.diag(cfs_matrix)
    FN = cfs_matrix.sum(axis=0) - np.diag(cfs_matrix)
    acc_each = TP/(TP+FN)
    
    #acc_list.append(acc_mean)
    df_mean[f"top{topk}_acc"] = acc_mean
    df_dict[f"top{topk}_acc"] = acc_each
```

```python
df_roc.to_csv(f"{result_dir}{predictor}_roc.csv", index = None)
df_mean.to_csv(f"{result_dir}{predictor}_mean.csv", index = None)
df_dict.to_csv(f"{result_dir}{predictor}_each.csv", index = None)
```
